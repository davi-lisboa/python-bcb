# SGS Economia Regional

Exemplos de como acessar as séries temporais da inadimplência das operações de crédito.

Disponível em: [SGS - Sistema Gerenciador de Séries Temporais - v2.1](https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries)

````python
from bcb.sgs.regional_economy import get_non_performing_loans, get_non_performing_loans_codes
from bcb.utils import BRAZILIAN_REGIONS, BRAZILIAN_STATES
import pandas as pd
````

## Busca por estado

No primeiro exemplo:

```
Economia regional
  - Norte
    - Roraima
      - Crédito
```

````python
series = get_non_performing_loans(["RR"], last=10, mode="all")
series
````

````python
series.plot()
````

## Busca por região

Nesse exemplo, últimos 5 registros de pessoa jurídica no Nordeste.

```
Economia regional
  - Nordeste
    - Consolidado regional
      - Crédito
```

````python
northeast_states = BRAZILIAN_REGIONS["NE"]
series = get_non_performing_loans(northeast_states, last=5, mode="pj")
series
````

## Buscando por todos os estados brasileiros

````python
series = get_non_performing_loans(BRAZILIAN_STATES, mode="PF", start="2024-01-01")
series
````

````python
series.plot()
````

## Exportando para CSV

````python
series.to_csv('inadimplencia_pf_por_estado_2024.csv')
````
